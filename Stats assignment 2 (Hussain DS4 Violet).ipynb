3. ### For the same columns Calculate IQR and identify outliers in these columns

### Q1 = df[column].quantile(0.25)

### Q3 = df[column].quantile(0.75)

### IQR = Q3 - Q1

### outliers = df[(df[column] < Q1 - 1.5 * IQR) | (df[column] > Q3 + 1.5 * IQR)]

```python
#IQR

Q1 = car_sales_data.quantile(0.25,numeric_only = True)
Q3 = car_sales_data.quantile(0.75,numeric_only = True)

IQR = Q3 - Q1
IQR.round(2)
```

```python
#Outliers

outliers = car_sales_data.copy()

#We will initiate a for loop to iterate through each column and check for outliers
for column in car_sales_data.select_dtypes(include="number").columns:
    outlier_conditions = (car_sales_data[column]< Q1[column] - 1.5 * IQR[column]) | (car_sales_data[column] > Q3[column] + 1.5 * IQR[column])
    
    #This line replaces values in the column of the outliers DataFrame with NaN where the corresponding 
    # values in the original car_sales_data DataFrame meet the outlier conditions; otherwise, it keeps the values unchanged.
    outliers[column]  = car_sales_data[column].where(~outlier_conditions)

#This will only show those columns that have numerical data
outliers = outliers.select_dtypes(include="number")
    
print(outliers)
```

4. ### Make Histograms for all Numerical columns and identify the distribution of type by calculating skewness;

### Note : 
- ### If the skewness is 0, it indicates a perfectly symmetrical distribution.
- ### If the skewness is negative, it means the distribution is skewed to the left (long left tail).
- ### If the skewness is positive, it means the distribution is skewed to the right (long right tail).

```python
#Collecting Numeric Columns
numerical_columns = car_sales_data.select_dtypes(include="number").columns

#Rows and columns
num_rows = len(numerical_columns) // 2
num_cols = 2

#Plot size 
fig, axes = plt.subplots(num_rows, num_cols, figsize= (15,3*num_rows))
fig.subplots_adjust (hspace= 0.5)

#For loop for histograms
for ax, cols in zip(axes.flatten(), numerical_columns):
    car_sales_data[cols].plot(kind = "hist", bins = 25, edgecolor = "k", ax=ax)
    ax.set_title (f"Histogram for {cols}")
    ax.set_xlabel (cols)
    ax.set_ylabel ("Frequency")
    
    #Now, we will calculate skewness
    skewness = car_sales_data[cols].skew()
   
    skewness_text = f'Skewness: {skewness:.2f}\n'

    #Here we are using and IF ELIF statement to print the Skewness type in the graph
    if skewness == 0:
        skewness_text += 'perfectly symmetrical.'
    elif skewness < 0:
        skewness_text += 'Negatively Skewed'
    else:
        skewness_text += 'Positively Skewed.'

    ax.text(0.95, 0.85, skewness_text, transform=ax.transAxes, ha='right', va='center', bbox=dict(facecolor='white', alpha=0.5))
plt.show()
```
